```python
# selenium으로 다나와 접속
# 웹 페이지의 HTML 정보 가져오기
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# selenium으로 다나와 검색 결과 URL에 접속

# colab에서 selenium을 사용 시 아래 코딩 작성 필요
# chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless') # headless : 내부 창을 띄우지 못하므로 아예 창을 띄우지 않게 설정
# chrome_options.add_argument('--no-sandbox')
# chrome_options.add_argument('--disable-dev-shm-usage')
# driver = webdriver.Chrome('chromedriver', options=chrome_options)
driver = webdriver.Chrome('C:/playwithdata/chromedriver.exe')
url = "http://search.danawa.com/dsearch.php?query=무선청소기&tab=main"
driver.get(url)
```

```python
# 주의 : 반드시 driver의 주소가 켜져있는 상태일 것
# 현재 페이지의 HTML 정보 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 1페이지에 대한 무선청소기 정보 가져오기 - 1
prod_items = soup.select("li.prod_item")
len(prod_items)
```

```python
# 1페이지에 대한 무선청소기 정보 가져오기 - 2
prod_items = soup.select('ul.product_list > li.prod_item')
len(prod_items)
```

```python
# 1페이지에 대한 무선청소기 정보 가져오기 - 3
prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
len(prod_items)
```

```python
# 상품명 정보 가져오기
title = prod_items[0].select('p.prod_name > a')[0].text.strip()
print(title)
```

```python
# 스펙 목록 정보 가져오기
spec_list = prod_items[0].select('div.spec_list')[0].text.strip()
print(spec_list)
```

```python
# 가격 정보 가져오기
price = prod_items[0].select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",","")
print(price)
```

```python
# 반복문으로 검색 결과의 1페이지에 대한 상품 정보 추출
prod_data = []
for prod_item in prod_items:
    try: # 상품명 가져오기
        title = prod_item.select('p.prod_name > a')[0].text.strip()
    except:
        title = ''
    try : # 스펙 목록 가져오기
        spec_list = prod_item.select('div.spec_list')[0].text.strip()
    except:
        spec_list = ''
    try : # 가격 정보 가져오기
        price = prod_item.select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",", "")
    except:
        price = 0
        
    prod_data.append([title, spec_list, price])
    
print(len(prod_data))
print(prod_data)
```

```python
# 상품 정보 태그에서 원하는 정보를 추출하는 함수
def get_prod_items(prod_items):
    prod_data = []
    for prod_item in prod_items:
        try: # 상품명 가져오기
            title = prod_item.select('p.prod_name > a')[0].text.strip()
        except:
            title = ''
        try : # 스펙 목록 가져오기
            spec_list = prod_item.select('div.spec_list')[0].text.strip()
        except:
            spec_list = ''
        try : # 가격 정보 가져오기
            price = prod_item.select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",", "")
        except:
            price = 0
        
        prod_data.append([title, spec_list, price])
    
    return prod_data
```

```python
# 상품 정보를 가져오는 함수 테스트
prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
prod_data = get_prod_items(prod_items)
print(len(prod_data))
```
