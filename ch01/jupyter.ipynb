```python
# pip3 install opencv-python
import cv2
import matplotlib.pyplot as plt
```

```python
# opencv 채널 순서는 b, g, r 순으로 되어있음
# 컬러 -> 흑백, 흑백 -> 컬러로 바꾸고 싶을 때 flags 설정
img = cv2.imread('cat.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
# 세로 x 가로 x 채녈(컬러)
img.shape
```

```python
cv2.imwrite('gray_cat.jpg', img, params=[cv2.IMWRITE_JPEG_QUALITY, 80])
```

```python
# opencv의 imshow함수 -> jupyter notebook애서는 잘 안씀 .py; 
# imshow에선 창이 별도로 뜨므로 matplotlib.pyplot.show()를 사용
cv2.imshow('win', img) 

# while True:
    
#     # 100ms동안 키입력을 기다림. ESC 키
#     if cv2.waitKey(100) == 27:
#         break
  
# 아무 키나 입력이 들어오면 창 닫고 종료    
cv2.waitKey()    

#창 닫기
cv2.destroyWindow('win')   
```

```python
# matplotlib은 채널의 순서를 r, g, b 순
# matplotlib으로 그림을 그리려면 bgr -> rgb로 바꿔야 함 : convert Color; cvtColor
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
